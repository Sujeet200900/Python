```python
summer=pd.read_csv('C://Users//HP//Documents//Basic_Pandas//summer.csv')
```

```python
summer
```

```python
summer1=summer[['Country','Medal','Gender']]
```

```python
summer2=summer1.groupby(['Country','Medal'])['Gender'].count().reset_index()
```

```python
summer2.rename(columns ={'Gender':'Count'},inplace=True)
```

```python
summer2
```

```python
summer2.pivot(index="Country", columns="Medal", values="Count").fillna(0)
```

```python
summer2.pivot_table(index="Country", columns="Medal", values="Count", aggfunc="mean").fillna(0)
```

```python
summer2.pivot_table(index="Country", columns="Medal", values="Count", aggfunc="sum",fill_value=0, margins=True)
```
